```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('./PCOS_data_without_infertility.xlsx', sheet_name='Full_new')
```

```python
df.head()
```

```python
print(df.isnull().sum() / df.shape[0])
```

```python
df.drop(['Sl. No', 'Patient File No.', 'Unnamed: 44'], axis=1, inplace=True)
```

```python
print(df.head())
```

```python
df.columns
```

```python
df['FSH/LH']
```

```python
for col in df.columns:
    print(f"{col}: {df[col].nunique()}")
```

```python
import numpy as np
```

```python
df['Blood Group'] = df['Blood Group'].apply(lambda x: ['A+', 'A-', 'B+', 'B-', 'O+', 'O-', 'AB+', 'AB-'][x-11])

cols = ['PCOS (Y/N)', 'Pregnant(Y/N)', 'Weight gain(Y/N)', 'hair growth(Y/N)', 'Skin darkening (Y/N)', 'Hair loss(Y/N)', 'Pimples(Y/N)', 'Fast food (Y/N)', 'Reg.Exercise(Y/N)']

for col in cols:
    df[col] = np.where(df[col] == 1, 'yes', np.where(df[col] == 0, 'no', None))
```

```python
df.shape[0]
```

```python
df = df.dropna()
```

```python
df.shape[0]
```

```python
df.head()
```

```python
df.dtypes
```

```python
cols = ['II    beta-HCG(mIU/mL)', 'AMH(ng/mL)']

df[cols] = df[cols].replace(r'[^0-9.\-]', None, regex=True)

df[cols] = df[cols].replace(r'\.$', None, regex=True)

df[cols] = df[cols].astype(float)
```

```python
cols = df.select_dtypes(exclude=['number']).columns
df[cols] = df[cols].astype('category')
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap((df.select_dtypes('number').corr() * 100).astype(int), annot=True)
plt.show()
```

```python
df.to_csv('pcos_dashboard.csv', index=False)
```
